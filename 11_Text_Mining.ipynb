# Perform sentimental analysis on the Elon-musk tweets

```python
import pandas as pd
import numpy as np
import nltk
import string
# 
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
from wordcloud import WordCloud, STOPWORDS
from nltk.sentiment.vader import SentimentIntensityAnalyzer
%matplotlib inline
```

```python
EM_sentiment = pd.read_csv('Elon_musk.csv', encoding='ISO-8859-1',index_col=0)
```

```python
EM_sentiment
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
def get_sentiment_scores(text):
    return sia.polarity_scores(text)
```

```python
EM_sentiment['Sentiment Scores'] = EM_sentiment['Text'].apply(get_sentiment_scores)
```

```python
EM_sentiment['Sentiment Score'] = EM_sentiment['Sentiment Scores'].apply(lambda x: x['compound'])
```

```python
def get_sentiment_category(sentiment_scores):
    if sentiment_scores['compound'] > 0.05:
        return 'Positive'
    elif sentiment_scores['compound'] < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

EM_sentiment['Sentiment Category'] = EM_sentiment['Sentiment Scores'].apply(get_sentiment_category)
```

```python
EM_sentiment.drop('Sentiment Scores', axis=1, inplace=True)
```

# Extract reviews of any product from ecommerce website like amazon
Perform emotion mining

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.amazon.in/Xiaomi-Storage-Snapdragon-Flagship-Cameras/product-reviews/B09XBCCQJT?reviewerType=all_reviews'
```

```python
url
```

```python
response = requests.get(url)
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
reviews = soup.find_all('span', {'data-hook': 'review-body'})
```

```python
sia = SentimentIntensityAnalyzer()

for review in reviews:
    text = review.text.strip()
    score = sia.polarity_scores(text)['compound']
    if score > 0.5:
        print(f"Positive review: {text}")
    elif score < 0.5:
        print(f"Negative review: {text}")
    else:
        print(f"Neutral review: {text}")
```

# Stopwords

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
import re
```

```python
para = pd.read_csv('stop.txt')
```

```python
para = ' '.join(para['a'].tolist())
```

```python
para
```

```python
len(para)
```

```python
para[0]
```

```python
sent = sent_tokenize(para)
sent
```

```python
sent[0]
```

```python
len(sent)
```

```python
corpus = []

for i in range(len(sent)):
    txt = re.sub('[^a-zA-Z]',' ',sent[i])
    txt = txt.lower()
    corpus.append(txt)
```

```python
corpus
```

```python
len(corpus)
```

```python
stemmer = PorterStemmer()

stemmer.stem('history')
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
for i in corpus:
    words = word_tokenize(i)
    for i in words:
        if i not in stopwords.words('english'):
            print(stemmer.stem(i))
```

```python
lemma = WordNetLemmatizer()

lemma.lemmatize('history')
```

```python
for i in corpus:
    words = word_tokenize(i)
    for i in words:
        if i not in stopwords.words('english'):
            print(lemma.lemmatize(i))
```
